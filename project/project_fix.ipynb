### Обработка выбросов и пропусков

```python
years_in_current_job_median_global = 0

def modify_data(dataframe, bankruptcies_median, credit_score_median, annual_income_median, years_in_current_job_median=False):
    df = dataframe.copy()

    df.loc[df['Bankruptcies'].isnull(), 'Bankruptcies'] = bankruptcies_median
    df.loc[df['Credit Score'] > 1000, ['Credit Score']] /= 10    
    df.loc[df['Credit Score'].isnull(), 'Credit Score'] = credit_score_median
    
    df.loc[df['Annual Income'].isnull(), 'Annual Income'] = annual_income_median
    df.loc[df['Months since last delinquent'].isnull(), 'Months since last delinquent'] = -1
    
    
    df.loc[df['Term'] == 'Short Term', 'Term'] = '0'
    df.loc[df['Term'] == 'Long Term', 'Term'] = '1'
    df['Term'] = df['Term'].astype("int8")
    
    # Заменяем Purpose, Home Ownership на dummy признаки
    df['Purpose'] = df['Purpose'].astype("category")
    df['Home Ownership'] = df['Home Ownership'].astype("category")
    df = pd.get_dummies(df, columns=['Home Ownership', 'Purpose'])
    
    # Обработка стажа
    
    df['Years in current job'].replace(to_replace=r'[^\d]+', value='', inplace=True, regex=True)
    df.loc[df['Years in current job'].isna() == True, 'Years in current job'] = '-1'
    df['Years in current job'] = df['Years in current job'].astype(int)
    if years_in_current_job_median:
        df.loc[df['Years in current job'] == 999, 'Years in current job'] = years_in_current_job_median
    else:
        global years_in_current_job_median_global
        years_in_current_job_median_global= int(df['Years in current job'].median())
        df.loc[df['Years in current job'] == 999, 'Years in current job'] = years_in_current_job_median_global
    
    
    return df
    

    
   
```

```python
def balance_df_by_target(df, target_name):

    target_counts = df[target_name].value_counts()

    major_class_name = target_counts.argmax()
    minor_class_name = target_counts.argmin()

    disbalance_coeff = int(target_counts[major_class_name] / target_counts[minor_class_name]) - 1

    for i in range(disbalance_coeff):
        sample = df[df[target_name] == minor_class_name].sample(target_counts[minor_class_name])
        df = df.append(sample, ignore_index=True)

    return df.sample(frac=1)
```

```python
def get_classification_report(y_train_true, y_train_pred, y_test_true, y_test_pred):
    print('TRAIN\n\n' + classification_report(y_train_true, y_train_pred))
    print('TEST\n\n' + classification_report(y_test_true, y_test_pred))
    print('CONFUSION MATRIX\n')
    print(pd.crosstab(y_test_true, y_test_pred))
```
