#### 1. Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

Precision это отношение верных "срабатываний" алгоритма к общему числу "срабатываний". Максимизировать его имеет смысл, когда нам нужны на выходе "чистые" данные, самые яркие объекты класса, ценой этого является то, что мы согласны "заплатить" тем что на некоторых объектах искомого класса алгоритм не сработает.
Применять эту метрику целесообразно, когда ошибки FN приемлемы, а FP - категорически не приемлемы.

Например, алгоритм выписывающий штрафы. В сомнительной ситуации лучше "промолчать", чем выписать заведомо неправомерный штраф, который может быть оспорен или коснется невиновного. 

Второй пример - поиск, скажем, рекомендованных фильмов/песен/книг - нет цели отобразить пользователю все экземпляры нужного класса из базы, но те несколько, что выводятся на страницу, должны максимально соответствовать классу предпочтений этого пользователя.

Recall - это мера, все ли правильные ответы нашел классификатор. При этом ошибки FP не учитываются, и в крайнем случае, относя все объекты к искомому классу, мы получим единицу recall, при отсутствии ценности предсказаний. Применима эта метрика когда более приемлемы FP ошибки.

Например, мошеннические операции по картам - перевести деньги злоумышленникам - значительно большая "цена ошибки", чем отклонить операцию.

Ну и классический пример с диагностикой - пропустить болезнь у пациента гораздо более дорогая ошибка, чем отправить здорового на дополнительное обследование.

В общем случае, recall стоит использовать на асиммитричных данных. Когда искомого класса объектов в процентном отношении мало, и тогда естественно, предпочтительно отыскать максимум из них.   

#### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Строго говоря, вопрос не корректен. Среднее от Precision и Recall можно взять тремя разными способами - арифметическое, геометрическое и гармоническое среднее. Последнее и является F-мерой.
Почему именно гармоническое? Оно хорошо подходит для усреднения мер. С нашем случае это Precision и Recall, но могут быть производительность, или, например, скорость. 

Со скоростью самый очевидный пример: Половина пути пройдена на машине со скоростью 100 км/ч, половина - пешком по болоту, со скоростью 2 км/ч. Очевидно, что при попытке применить среднее арифметическое, мы получим неверный результат (51км/ч) , потому что не будет учтено что время в пути на автомобиле в разы меньше, чем пешком. А реальная средняя скорость - 3.92 км/ч. 

Аналогично при оценке алгоритма, если одна из метрик, Precision или Recall близка к единице, а вторая - к нулю, среднее арифметическое будет возле 0.5, что не особо информативно. У F-меры же в числителе $P*R$, и если одна из метрик будет мала, тогда и F-мера тоже будет мала, мы увидим что в среднем, наш алгоритм ошибается часто.

Конечно, для того чтобы понимать какого типа ошибки, полезно выводить и анализировать все три меры сразу, причем для всех возможных классов по отдельности, т.к. не факт, что алгоритм, безошибочно определяющий "котиков" справится с "рыбками".
