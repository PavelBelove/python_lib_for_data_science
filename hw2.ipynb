#### 1. В чём различие между зависимыми и независимыми выборками?

Независимые выборки, как следует из названия, возникают при сравнивании двух разных групп. Выбор объекта в одной из групп никак не влияет на другую. Примеры: молодые и старые. Мужчины и женщины. Автомобили "Тойота" и автомобили "Ауди".
В зависимых выборках выбор объекта в одной выборке влияет на выбор в другой. Примеры: Одна и та же выборка до эксперимента и после. Мужья и их жены. Близнецы. 

#### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические критерии "мощнее", но в основе их применения лежит целый набор допущений, которым должны удовлетворять исследовательские данные (например, форма распределения выборочных статистик, равенство дисперсий) для того, чтобы соответствующий критерий можно было использовать.
Поэтому если данные подходят для применения параметрического критерия и есть возможность применить его - выбираем параметрический.
Если же, например, форма распределения далека от нормального - используем непараметрические критерии, не такие чувствительные к "чистоте" входящих данных.
